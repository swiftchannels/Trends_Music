```python
#Now lets try another API  method
#Let's go to https://www.last.fm/api/show/album.search
#There are two choices to go about requests but lets use JSON
# So it would be  /2.0/?method=album.search&album=believe&api_key=YOUR_API_KEY&format=json

#Let's attach the url within the requests.get() call

album_search = 'believe' #Search for the album...
response = requests.get('http://ws.audioscrobbler.com/2.0/?method=album.search&album=' + album_search +  '&api_key='  + api_key +'&format=json')
json_data = json.loads(response.text) #The best way to view API data response is via JSON so we could sepearate them and get the data we only need
print(json_data)

#See http://prntscr.com/nz7zly to view API heirarchy order on data
results = json_data['results'] #First level
albums = results['albummatches'] #what we are interested in getting
albums = albums['album'] #The level that contains the data that we need
#The results text should be logging every dictionary objects under 'album'


```

```python
#Let's get the name of the album, the artist and the url
for album in albums:
    print('Name ' + str(album['name'])) #Remember to turn your json object into string before concatenating
    print('Aritst ' + str(album['artist']))
    print('Url ' + str(album['url']))
```
